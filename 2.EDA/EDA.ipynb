```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/dataset_student_depression_cleaned')
```

```python
sorted(df['Study Hours'].astype(str).unique())
```

```python
df["Study Hours"] = df["Study Hours"].astype(float).astype(int)
```

```python
df
```

```python
null_count = df.isnull().sum()
print(null_count)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
age_dep_counts = df.groupby(["Age Category", "Depression"]).size().reset_index(name="Count")

# Plot
plt.figure(figsize=(8,5))
sns.barplot(data=age_dep_counts, x="Age Category", y="Count", hue="Depression", palette="Set2")

plt.title("Depression by Age Category")
plt.ylabel("Number of Students")
plt.xlabel("Age Category")
plt.legend(title="Depression")
plt.show()
```

```python
age_dep_counts = df.groupby(["Age Category", "Depression"]).size().unstack(fill_value=0)

# Plot
age_dep_counts.plot(kind="bar", figsize=(8,5), color=["skyblue", "salmon"])

plt.title("Depression by Age Category")
plt.ylabel("Number of Students")
plt.xlabel("Age Category")
plt.xticks(rotation=0)
plt.legend(title="Depression")
plt.show()
```

```python
sleep_dep_counts = df.groupby(["Sleep Duration", "Depression"]).size().unstack(fill_value=0)

# Plot
sleep_dep_counts.plot(kind="bar", figsize=(8,5), color=["skyblue", "salmon"])

plt.title("Depression by Sleep Duration")
plt.ylabel("Number of Students")
plt.xlabel("Sleep Duration")
plt.xticks(rotation=45)
plt.legend(title="Depression")
plt.tight_layout()
plt.show()
```
