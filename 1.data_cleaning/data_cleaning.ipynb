```python
df_selected["Financial Stress"] = pd.to_numeric(df_selected["Financial Stress"], errors="coerce")
df_selected["Financial Stress"] = df_selected["Financial Stress"].apply(lambda x: "No" if x <= 3.0 else "Yes")

# Check result
print(df_selected["Financial Stress"].value_counts())
```

```python
df_selected.rename(columns={"Work/Study Hours": "Study Hours"}, inplace=True)
```

```python
print(df_selected["Sleep Duration"].value_counts())
```

```python
# Remove rows where Sleep Duration is 'Others'
df_selected = df_selected[df_selected["Sleep Duration"] != "Others"]

# Reset index after removal
df_selected.reset_index(drop=True, inplace=True)

```

```python
print(df_selected["Sleep Duration"].value_counts())
```

```python
# Remove quotes and extra spaces first
df_selected["Sleep Duration"] = df_selected["Sleep Duration"].str.replace("'", "").str.strip()

# Map values according to your rule
mapping = {
    "Less than 5 hours": 5,
    "5-6 hours": 6,
    "7-8 hours": 7,
    "More than 8 hours": "8+"
}

df_selected["Sleep Duration"] = df_selected["Sleep Duration"].map(mapping)

print(df_selected["Sleep Duration"].value_counts())
```

```python
df_selected
```

```python
print(df_selected["Age"].value_counts())
```

```python
# Make sure Age is numeric
df_selected["Age"] = pd.to_numeric(df_selected["Age"], errors="coerce")

# Define bins and labels
bins = [0, 18, 25, 30, float('inf')]
labels = ["0-18", "19-25", "26-30", "30+"]

# Create Age Category column
df_selected["Age Category"] = pd.cut(df_selected["Age"], bins=bins, labels=labels, right=True)

print(df_selected[["Age", "Age Category"]].head())
```
